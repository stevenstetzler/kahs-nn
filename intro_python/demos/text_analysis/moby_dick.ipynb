```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
in_file = open("moby_dick.txt", 'rb')
```

```python
text_lines = in_file.readlines()
text_lines
```

```python
decoded_text_lines = [line.decode('utf-8') for line in text_lines]
decoded_text_lines
```

```python
story = decoded_text_lines[849:21971]
story
```

```python
story_no_n = [line.strip('\n') for line in story]
story_no_n
```

```python
story_no_r = [line.rstrip('\r') for line in story_no_n]
story_no_r
```

```python
story_words_only = [line for line in story_no_r if len(line) > 0]
story_words_only
```

```python
story_processed = [line.lower() for line in story_words_only]
story_processed
```

```python
first_line = story_processed[0]
first_words = first_line.split()
first_words
```

```python
words = []
for line in story_processed:
    new_words = line.split()
    for new_word in new_words:
        words.append(new_word)
words
```

```python
print("Moby Dick is {} words long!".format(len(words)))
```

```python
words_to_instances = {}
for word in words:
    words_to_instances[word] = 0
words_to_instances['the']
```

```python
for word in words:
    number_instances = words_to_instances[word]
    words_to_instances[word] = number_instances + 1
```

```python
words_to_instances['the']
```

```python
words_to_instances['whale']
```

```python
unique_words = list(words_to_instances.keys())
unique_words
```

```python
print("There are {} unique words in Moby Dick!".format(len(unique_words)))
```

```python
import operator
sorted_words = sorted(words_to_instances.items(), key=operator.itemgetter(1), reverse=True)
sorted_words
```

```python
N = 10
words_to_plot = sorted_words[0:10]
words_to_plot
```

```python
labels = [word_to_plot[0] for word_to_plot in words_to_plot]
values = [word_to_plot[1] for word_to_plot in words_to_plot]
print(labels)
print(values)
```

```python
x_axis = np.arange(N)
plt.bar(x_axis, values)
plt.xticks(x_axis, labels)
plt.show()
```

```python
np_words = np.asarray(words)
his_locations = np.where(np_words == 'his')[0]
his_locations
```

```python
x_axis = his_locations
y_axis = np.linspace(0, len(his_locations), len(his_locations))
print(y_axis)
plt.plot(x_axis, y_axis)
plt.xlabel("Word in Moby Dick")
plt.show()
```

```python
whale_locations = np.where(np_words == 'whale')[0]
x_axis = whale_locations
y_axis = np.linspace(0, len(whale_locations), len(whale_locations))
plt.plot(x_axis, y_axis)
plt.xlabel("Word in Moby Dick")
plt.show()
```

We can compare distributions of words to determine their importance. "His" is not an important word - it appears uniformly, but "whale" might be.
