# Getting the Data
We open text files simply using the `open(file_name)` function. Here, `'rb'` means "read in byte encoded data".

```python
in_file = open("moby_dick.txt", 'rb')
```

Next, we use the `readlines()` function to read the text file and pull in all of the lines of text. A "line" of text is simply text followed by the new line character `\n`, which is entered into a text file whenever you hit the 'Enter' or 'Return' keys.

```python
text_lines = in_file.readlines()
text_lines
```

Since we had to read in the data with byte encoding, we want to convert this data to strings using the "UTF-8" encoding, which is a standard encoding for text data. We use a **list concatenation** to apply the **string's** `decode()` function to each of the lines of our text.

```python
decoded_text_lines = [line.decode('utf-8') for line in text_lines]
decoded_text_lines
```

If you look at the accompanying `moby_dick.txt` text file, you'll see that there's a whole bunch of irrelevant information at the beginning and the end of the text file. We can get rid of it by using a **slice** operation on our list of lines. We use the syntax `sliced_data = data[start:end]`, so `story = decoded_text_lines[start_line_number:end_line_number]`.

```python
story = decoded_text_lines[849:21971]
story
```

We can do some preprocessing using the **string's** `strip()` command to remove the newlines (`\n`) and carriage returns (`\r`) encoded in the text.

```python
story_no_n = [line.strip('\n') for line in story]
story_no_r = [line.rstrip('\r') for line in story_no_n]
story_no_r
```

We don't want any lines that don't contain any text! We can find the **length** of a **string** or **list** using the `len()` function.

```python
story_words_only = [line for line in story_no_r if len(line) > 0]
story_words_only
```

And finally, we don't really care about the capitalization of words, so we can convert all of our words to lowercase using the **string's** `lower()` function.

```python
story_processed = [line.lower() for line in story_words_only]
story_processed
```

Now, we want to extract the individual words from our lines and put them into their own list. We can do this using the **string's** `split()` function, which *splits* a string on any whitespace in the string producing a list of the resulting individual components of the string. For example `"this is a line".split()` $\rightarrow$ `["this", "is", "a", "line"]`.

```python
first_line = story_processed[0]
first_words = first_line.split()
first_words
```

```python
words = []
for line in story_processed:
    new_words = line.split()
    for new_word in new_words:
        words.append(new_word)
words
```

Now we can find how many words there are in Moby Dick!

```python
print("Moby Dick is {} words long!".format(len(words)))
```

Now what if we want to keep track of **how often** each word appears in Moby Dick? Say we want to see how often the word "whale" is used in comparison to "moby dick". We can use a **dictionary** to keep track of this. First, we initialize our dictionary.

```python
words_to_instances = {}
for word in words:
    words_to_instances[word] = 0
words_to_instances['the']
```

Next, we loop over all of our words, and increment our dictionary's counter for how many times that word appeared in the text.

```python
for word in words:
    number_instances = words_to_instances[word]
    words_to_instances[word] = number_instances + 1
```

```python
words_to_instances['the']
```

```python
words_to_instances['whale']
```

```python
words_to_instances['moby']
```

The dictionary we created contains **unique** key values, which allows us to easily find how many unique words there are in Moby Dick.

```python
unique_words = list(words_to_instances.keys())
unique_words
```

```python
print("There are {} unique words in Moby Dick!".format(len(unique_words)))
print("{0:0.2f}% of words in Moby Dick are unique!".format(100. * len(unique_words)/len(words)))
```

We can also **sort** our dictionary to find which words were used most commonly. For the following line, I almost never remember how to sort items, so I looked it up [here](https://stackoverflow.com/questions/613183/how-do-i-sort-a-dictionary-by-value).

```python
import operator
sorted_words = sorted(words_to_instances.items(), key=operator.itemgetter(1), reverse=True)
sorted_words
```

```python
most_common_word, occurances = sorted_words[0]
print("The most common word in Moby Dick is {}, appearing {} times.".format(most_common_word, occurances))
print("The word '{}' accounts for {:0.2f}% of Moby Dick.".format(most_common_word, 100. * occurances / len(words)))
```

'the', 'of', and 'and' are the most commonly used words in Moby Dick. And the word 'the' alone takes up a large majority of the text... We can also represent this information graphically using the **pyplot** library.

```python
import numpy
import matplotlib.pyplot as pyplot
%matplotlib inline
```

```python
N = 10
words_to_plot = sorted_words[0:10]
words_to_plot
```

```python
labels = [word_to_plot[0] for word_to_plot in words_to_plot]
values = [word_to_plot[1] for word_to_plot in words_to_plot]
print(labels)
print(values)
```

```python
x_axis = numpy.arange(N)
pyplot.bar(x_axis, values)
pyplot.xticks(x_axis, labels)
pyplot.show()
```

We can also analyze a word's evolution over the course of the book, which can answer a lot of questions. For example, is a certain word used a lot in the beginning of the text and then never again? We will use the **numpy** and **pyplot** packages to do this. The **numpy** packaged uses **arrays** instead of **lists**. They aren't much different from lists, in that they can be indexed and worked with in similar ways. The function `np.where` gives us the **indices** where a certain value appears in a numpy array.

```python
array_words = numpy.asarray(words)
his_locations = numpy.where(array_words == 'his')[0]
his_locations
```

```python
x_axis = his_locations
y_axis = numpy.linspace(0, len(his_locations), len(his_locations))

print(y_axis)

pyplot.plot(x_axis, y_axis)

pyplot.title("The occurances of 'his' in Moby Dick")
pyplot.ylabel("Occurance number")
pyplot.xlabel("Word in Moby Dick")

pyplot.show()
```

```python
whale_locations = numpy.where(array_words == 'whale')[0]

x_axis = whale_locations
y_axis = numpy.linspace(0, len(whale_locations), len(whale_locations))

pyplot.plot(x_axis, y_axis)

pyplot.title("The occurances of 'whale' in Moby Dick")
pyplot.ylabel("Occurance number")
pyplot.xlabel("Word in Moby Dick")

pyplot.show()
```

We already have an interesting result here! "his" doesn't appear to be very interesting, since it appears uniformly throughout the text. But the word "whale" seems to appear infrequently at the beginning of the story, consistently during the middle of the story, and drops off again in frequency by the end.
